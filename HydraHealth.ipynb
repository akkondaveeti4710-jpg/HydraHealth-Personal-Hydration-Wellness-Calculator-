```python
import gradio as gr

# ----- BMI -----
def calculate_bmi(weight_kg, height_cm):
    height_m = height_cm / 100
    bmi = weight_kg / (height_m ** 2)
    return round(bmi, 2)

def bmi_category(bmi):
    if bmi < 18.5:
        return "Underweight"
    elif bmi < 25:
        return "Normal weight"
    elif bmi < 30:
        return "Overweight"
    else:
        return "Obese"

# ----- Hydration -----
def recommended_water(weight_kg, activity_level):
    base = weight_kg * 35 / 1000
    if activity_level == "low":
        return round(base, 2)
    elif activity_level == "moderate":
        return round(base + 0.5, 2)
    elif activity_level == "high":
        return round(base + 1.0, 2)
    else:
        return round(base, 2)

def hydration_status(intake_liters, recommended_liters):
    percent = intake_liters / recommended_liters
    if percent < 0.5:
        return "Severe dehydration risk"
    elif percent < 0.8:
        return "Moderate dehydration risk"
    elif percent < 1:
        return "Mild dehydration risk"
    else:
        return "Well hydrated"

# ----- Calories -----
def estimate_calories(age, weight, height, gender, activity_level):
    if gender.lower() == "male":
        bmr = 10 * weight + 6.25 * height - 5 * age + 5
    else:
        bmr = 10 * weight + 6.25 * height - 5 * age - 161

    if activity_level == "low":
        calories = bmr * 1.2
    elif activity_level == "moderate":
        calories = bmr * 1.55
    elif activity_level == "high":
        calories = bmr * 1.9
    else:
        calories = bmr * 1.2

    return round(calories)

# ----- Health Advice -----
def get_health_advice(bmi, hydration_status):
    advice = []
    if bmi < 18.5:
        advice.append("Consider increasing calorie intake with nutritious foods.")
    elif bmi >= 25 and bmi < 30:
        advice.append("Regular exercise and balanced diet may help maintain a healthy weight.")
    elif bmi >= 30:
        advice.append("Consult a healthcare professional for weight management guidance.")

    if hydration_status == "Severe dehydration risk":
        advice.append("Increase water intake immediately.")
    elif hydration_status == "Moderate dehydration risk":
        advice.append("Drink more water throughout the day.")
    elif hydration_status == "Mild dehydration risk":
        advice.append("Try to drink at least one extra glass of water.")
    else:
        advice.append("Great job staying hydrated.")

    advice.append("Aim for at least 7–9 hours of sleep per night.")
    advice.append("Regular physical activity improves overall health.")
    return advice

# ----- Main Function -----
def generate_health_report(name, age, gender, height, weight, activity_level, water_intake):
    bmi = calculate_bmi(weight, height)
    bmi_cat = bmi_category(bmi)
    recommended = recommended_water(weight, activity_level)
    hydration = hydration_status(water_intake, recommended)
    calories = estimate_calories(age, weight, height, gender, activity_level)
    advice_list = get_health_advice(bmi, hydration)

    report = f"""
----- HEALTH REPORT -----

Name: {name}
Age: {age}

BMI: {bmi}
BMI Category: {bmi_cat}

Recommended Water Intake: {recommended} L
Your Water Intake: {water_intake} L
Hydration Status: {hydration}

Estimated Daily Calories: {calories}

Health Advice:
"""
    for tip in advice_list:
        report += f"- {tip}\n"

    return report

# ----- Gradio Interface -----
iface = gr.Interface(
    fn=generate_health_report,
    inputs=[
        gr.Textbox(label="Name"),
        gr.Number(label="Age"),
        gr.Dropdown(label="Gender", choices=["male", "female"]),
        gr.Number(label="Height (cm)"),
        gr.Number(label="Weight (kg)"),
        gr.Dropdown(label="Activity Level", choices=["low", "moderate", "high"]),
        gr.Number(label="Daily Water Intake (L)")
    ],
    outputs=gr.Textbox(label="Health Report", lines=20),
    title="HydraHealth: Personal Hydration & Wellness Calculator",
    description="Calculate your BMI, hydration status, estimated daily calories, and get personalized health advice."
)

# ----- Launch for Hugging Face -----
iface.launch(server_name="0.0.0.0", server_port=7860, share=True)
```
